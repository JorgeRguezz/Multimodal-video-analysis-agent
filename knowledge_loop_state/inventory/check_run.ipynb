```python
import json
from pathlib import Path

data = json.loads(Path("latest_inventory.json").read_text())

print("profile:", data["loop_profile_id"])
print("videos:", len(data["videos"]))
print("topic/facet buckets:", len(data["topic_facets"]))
print("warnings:", len(data["warnings"]))

for row in sorted(data["topic_facets"], key=lambda r: r["coverage_score"], reverse=True)[1000:1100]:
    print(row["topic_key"], row["facet"], row["coverage_score"], row["video_count"], row["chunk_count"])
```
